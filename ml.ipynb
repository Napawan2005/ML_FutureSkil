#**เช็คตัวแปลทีละตัวแปลว่าข้อมูลเป็นลักษณะยังไง [Univariate analysis]**

# **--- ข้อมูลส่วนตัว ---**

```person_age``` = อายุผู้กู้<br>

```person_income``` = รายได้ต่อปี<br>

```person_home_ownership```= สถานะการครอบครองที่อยู่อาศัย *เช่า(Rent) , เป็นเจ้าของ(Own) , กำลังผ่อนจำนอง(Mortgage)*

```person_emp_length``` = อายุงานที่ทำงานปัจจุบัน เพื่อวัดความมั่งคงอาชีพ

# **---ข้อมูลเกี่ยวกับสินเชื่อ ---**

```loan_intent``` = กู้เงินไปทำอะไร <br>
```loan_grade``` = เกรดคุณภาพหนี้ที่ธนาคารจัดลำดับให้ (เช่น A, B, C, D)<br>
```loan_amnt``` = จำนวนเงินที่ขอกู้<br> (ยิ่งยอดกู้สูง ภาระผ่อนต่อเดือนก็สูง ความเสี่ยงก็จะเพิ่มขึ้นตาม)

```loan_int_ratre``` = อัตราดอกเบี้ย (มักแปรผันตามความเสี่ยง[Risk-based procing])


#**--- ประวัติเครดิต ---**
```cb_person_default_on_file``` = อดีตที่ผิดนัดชำระหนี้ (Y/N)[ชี้ประวัติพฤติกรรมของลูกค้า]

```cd_person_cred_hist_leangth``` = ความยาวประวัติเครดิต

# **show plot good bad**

````python
df_target = df[['loan_status']]
df_target['loan_status'] = np.where(df_target['loan_status']==1 , 'Bad' , 'Good')
````

````python
df_target['loan_status']
````

# **สัดส่วนข้อมูลของขอ้มูลใน DataSet**

````python
print(df_target['loan_status'].value_counts())
print('*'*40)
print(df_target['loan_status'].value_counts() / len(df_target)*100)
print('*'*40)
````

# **สร้างกราฟ**

````python
fig = plt.figure(figsize=(8,6))
sns.countplot(x = 'loan_status' , data=df_target , palette = 'tab10')
plt.xlabel('Target')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
````

## เช็คจำนวน bad  , good

````python
count_good = len(df_target.loc[df_target['loan_status'] == 'Good'])
count_bad = len(df_target.loc[df_target['loan_status'] == 'Bad'])
````

````python
print("count good : ", count_good)
print("count bad : ",count_bad)
````
